```python
%%writefile app.py 
# Creamos el archivo de la APP en el interprete principal (Python)

############################# IMPLEMENTACIÓN DE DASHBOARD ################################

# Importamos librerías
from funpymodeling.exploratory import freq_tbl 
import streamlit as st
import plotly.express as px
import pandas as pd
import seaborn as sns  # Importa seaborn para los gráficos
import matplotlib.pyplot as plt  # Importa matplotlib para mostrar el gráfico de calor
from sklearn.linear_model import LinearRegression
import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
#################################################################
@st.cache_resource
def load_data():
    df1 = pd.read_csv("Dublin.csv")
    return df1

df1 = load_data()
##########################################################################################
# Agregar imagen en la barra lateral
image = "https://seeklogo.com/images/D/dublin-logo-30EF152DC2-seeklogo.com.png"
st.sidebar.image(image, use_column_width=True)


###################### CREACIÓN DEL DASHBOARD ################################

st.sidebar.title("DASHBOARD_Dublin")
st.sidebar.header("Sidebar")

Frames = st.sidebar.selectbox(label="Frames", options=["Análisis Univariado Dublin", "Regresión lineal Simple", "Regresión Múltiple", "Regresión no lineal"])
#########################################################################
if Frames == "Análisis Univariado Dublin":
    st.title("Análisis Univariado de Variables Categóricas Dublin")
    
    if st.sidebar.checkbox("Mostrar Dataset"):
        st.write(df1)

    Variables = st.sidebar.selectbox("Variable categórica", df1.select_dtypes(include=['object']).columns)
    Vars_Num = st.selectbox("Métrica a graficar", df1.select_dtypes(include=['float', 'int']).columns)
    tipo_grafico = st.sidebar.selectbox("Tipo de gráfico", ["Barras", "Líneas", "Dispersión", "Circular"])
    vars_multiselect = st.sidebar.multiselect("Selecciona múltiples métricas a graficar", df1.select_dtypes(include=['float', 'int']).columns)

    if st.sidebar.button("Actualizar Gráfico"):
        if tipo_grafico == "Barras":
            figure = px.bar(data_frame=df1, x=Variables, y=Vars_Num)
        elif tipo_grafico == "Líneas":
            figure = px.line(data_frame=df1, x=Variables, y=Vars_Num)
        elif tipo_grafico == "Dispersión":
            figure = px.scatter(data_frame=df1, x=Variables, y=Vars_Num)
        elif tipo_grafico == "Circular":
            figure = px.pie(data_frame=df1, names=Variables)

        st.plotly_chart(figure)

        for var in vars_multiselect:
            if tipo_grafico == "Barras":
                fig_multi = px.bar(data_frame=df1, x=Variables, y=var)
            elif tipo_grafico == "Líneas":
                fig_multi = px.line(data_frame=df1, x=Variables, y=var)
            elif tipo_grafico == "Dispersión":
                fig_multi = px.scatter(data_frame=df1, x=Variables, y=var)
            elif tipo_grafico == "Circular":
                fig_multi = px.pie(data_frame=df1, names=Variables)

            st.plotly_chart(fig_multi)
#########################################################################
if Frames == "Regresión lineal Simple":
    st.title("Regresión lineal Simple Dublin")

    # Selección de variables independientes y dependientes
    vars_indep = st.multiselect("Selecciona variables independientes", df1.select_dtypes(include=['float', 'int']).columns)
    var_dep = st.selectbox("Selecciona variable dependiente", df1.select_dtypes(include=['float', 'int']).columns)

    # Checkbox para mostrar el conjunto de datos
    if st.checkbox("Mostrar Dataset"):
        st.write(df1)

    # Botón para realizar la regresión
    if st.button("Ejecutar Regresión"):
        # Verificamos que se hayan seleccionado variables
        if len(vars_indep) == 0:
            st.warning("Por favor, selecciona al menos una variable independiente.")
        else:
            # Definimos las variables para la regresión
            Vars_Indep = df1[vars_indep]
            Var_Dep = df1[var_dep]

            # Limpiamos el DataFrame para la correlación
            df_clean = df1[vars_indep + [var_dep]].dropna()  # Solo seleccionamos las columnas necesarias y eliminamos nulos

            # Importamos la clase de regresión lineal
            from sklearn.linear_model import LinearRegression
            model = LinearRegression()

            # Ajustamos el modelo
            model.fit(X=Vars_Indep, y=Var_Dep)

            # Mostramos los coeficientes
            coef = model.coef_
            intercept = model.intercept_
            st.write("Coeficientes:", coef)
            st.write("Intercepto:", intercept)

            # Evaluamos el modelo
            r_squared = model.score(Vars_Indep, Var_Dep)
            st.write("R²:", r_squared)

            # Predicciones
            y_pred = model.predict(X=Vars_Indep)
            df1['Predicciones'] = y_pred

            # Gráfico de dispersión
            fig = px.scatter(df1, x=vars_indep[0], y=var_dep, title='Gráfico de Dispersión')
            st.plotly_chart(fig)

            # Gráfico de calor de correlación
            Corr_Factors1 = df_clean.corr()
            Heat_Map = sns.heatmap(Corr_Factors1, cmap='coolwarm', annot=True)
            st.pyplot(fig=Heat_Map.figure)
###########################################################################################
if Frames == "Regresión Múltiple":
    st.title("Regresión Múltiple Dublin")

    # Selección de variables independientes y dependientes
    vars_indep = st.multiselect("Selecciona variables independientes", df1.select_dtypes(include=['float', 'int']).columns)
    var_dep = st.selectbox("Selecciona variable dependiente", df1.select_dtypes(include=['float', 'int']).columns)

    # Checkbox para mostrar el conjunto de datos
    if st.checkbox("Mostrar Dataset"):
        st.write(df1)

    # Botón para realizar la regresión
    if st.button("Ejecutar Regresión"):
        # Verificamos que se hayan seleccionado variables
        if len(vars_indep) == 0:
            st.warning("Por favor, selecciona al menos una variable independiente.")
        else:
            # Definimos las variables para la regresión
            Vars_Indep = df1[vars_indep]
            Var_Dep = df1[var_dep]

            # Limpiamos el DataFrame para la correlación
            df_clean = df1[vars_indep + [var_dep]].dropna()  # Solo seleccionamos las columnas necesarias y eliminamos nulos

            # Ajustamos el modelo de regresión múltiple
            model = LinearRegression()
            model.fit(X=Vars_Indep, y=Var_Dep)

            # Mostramos los coeficientes
            coef = model.coef_
            intercept = model.intercept_
            st.write("Coeficientes:", coef)
            st.write("Intercepto:", intercept)

            # Evaluamos el modelo
            r_squared = model.score(Vars_Indep, Var_Dep)
            st.write("R²:", r_squared)

            # Predicciones
            y_pred = model.predict(X=Vars_Indep)
            df1['Predicciones'] = y_pred

            # Gráficos de dispersión para cada variable independiente con color
            for var in vars_indep:
                fig = px.scatter(df1, x=var, y=var_dep,
                                 title=f'Gráfico de Dispersión: {var} vs {var_dep}',
                                 color=var_dep,  # Colorear según la variable dependiente
                                 color_continuous_scale=px.colors.sequential.Viridis,  # Paleta de colores
                                 size_max=10)  # Ajustar el tamaño máximo de los puntos
                st.plotly_chart(fig)

                # Agregar la línea de regresión al gráfico de dispersión
                fig.add_traces(px.scatter(df1, x=var, y=y_pred, trendline="ols").data)

            # Gráfico de calor de correlación
            Corr_Factors1 = df_clean.corr()
            plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura
            Heat_Map = sns.heatmap(Corr_Factors1, cmap='coolwarm', annot=True)
            st.pyplot(fig=Heat_Map.figure)
##################################################################################
# Definir funciones no lineales
def func_quadratic(x, a, b, c):
    return a * x**2 + b * x + c

def func_exponential(x, a, b, c):
    return a * np.exp(b * x) + c

def func_inverse(x, a):
    return 1 / a * x

def func_sine(x, a, b):
    return a * np.sin(x) + b

def func_tangent(x, a, b):
    return a * np.tan(x) + b

def func_absolute(x, a, b, c):
    return a * np.abs(x) + b * x + c

def func_polynomial_ratio(x, a, b, c):
    return (a * x**2 + b) / (c * x)

def func_logarithmic(x, a, b):
    return a * np.log(x) + b

def func_linear_product(x, a, b, c):
    return a * x + b * x + c * x

def func_inverse_quadratic(x, a):
    return 1 / a * x**2

def func_inverse_polynomial(x, a, b, c):
    return a / b * x**2 + c * x

def mostrar_ecuacion(funcion, parametros):
    if funcion == "Función cuadrática":
        return f"y = {parametros[0]:.4f}x² + {parametros[1]:.4f}x + {parametros[2]:.4f}"
    elif funcion == "Función exponencial":
        return f"y = {parametros[0]:.4f}e^({parametros[1]:.4f}x) + {parametros[2]:.4f}"
    elif funcion == "Función inversa":
        return f"y = 1 / ({parametros[0]:.4f}) * x"
    elif funcion == "Función senoidal":
        return f"y = {parametros[0]:.4f}sin(x) + {parametros[1]:.4f}"
    elif funcion == "Función tangencial":
        return f"y = {parametros[0]:.4f}tan(x) + {parametros[1]:.4f}"
    elif funcion == "Función valor absoluto":
        return f"y = {parametros[0]:.4f}|x| + {parametros[1]:.4f}x + {parametros[2]:.4f}"
    elif funcion == "Función cociente entre polinomios":
        return f"y = ({parametros[0]:.4f}x² + {parametros[1]:.4f}) / ({parametros[2]:.4f}x)"
    elif funcion == "Función logarítmica":
        return f"y = {parametros[0]:.4f}ln(x) + {parametros[1]:.4f}"
    elif funcion == "Función lineal con producto de coeficientes":
        return f"y = {parametros[0]:.4f}x + {parametros[1]:.4f}x + {parametros[2]:.4f}x"
    elif funcion == "Función cuadrática inversa":
        return f"y = 1 / ({parametros[0]:.4f}) * x²"
    elif funcion == "Función polinomial inversa":
        return f"y = ({parametros[0]:.4f} / {parametros[1]:.4f}) * x² + {parametros[2]:.4f}x"

if Frames == "Regresión no lineal":
    st.title("Regresión no lineal Dublin")

    # Selección de variables independientes y dependientes
    x_variable = st.selectbox('Selecciona la variable independiente (x)', df1.select_dtypes(include=['float', 'int']).columns)
    y_variable = st.selectbox('Selecciona la variable dependiente (y)', df1.select_dtypes(include=['float', 'int']).columns)

    # Asegúrate de que los datos seleccionados no tengan valores NaN
    Vars_Indep = df1[[x_variable]].dropna()
    Var_Dep = df1[y_variable].dropna()

    # Redefinir las variables para el ajuste
    x = Vars_Indep[x_variable].values
    y = Var_Dep.values

    # Crear un selector para elegir la función de ajuste
    funcion_seleccionada = st.selectbox('Selecciona la función de ajuste', 
        ["Función cuadrática", "Función exponencial", "Función inversa", 
         "Función senoidal", "Función tangencial", "Función valor absoluto", 
         "Función cociente entre polinomios", "Función logarítmica", 
         "Función lineal con producto de coeficientes", 
         "Función cuadrática inversa", "Función polinomial inversa"])

    # Ajustar el modelo según la función seleccionada
    try:
        if funcion_seleccionada == "Función cuadrática":
            parametros, _ = curve_fit(func_quadratic, x, y)
            yfit = func_quadratic(x, *parametros)
        elif funcion_seleccionada == "Función exponencial":
            parametros, _ = curve_fit(func_exponential, x, y, p0=[1, 0.1, 1])
            yfit = func_exponential(x, *parametros)
        elif funcion_seleccionada == "Función inversa":
            parametros, _ = curve_fit(func_inverse, x, y)
            yfit = func_inverse(x, *parametros)
        elif funcion_seleccionada == "Función senoidal":
            parametros, _ = curve_fit(func_sine, x, y)
            yfit = func_sine(x, *parametros)
        elif funcion_seleccionada == "Función tangencial":
            parametros, _ = curve_fit(func_tangent, x, y)
            yfit = func_tangent(x, *parametros)
        elif funcion_seleccionada == "Función valor absoluto":
            parametros, _ = curve_fit(func_absolute, x, y)
            yfit = func_absolute(x, *parametros)
        elif funcion_seleccionada == "Función cociente entre polinomios":
            parametros, _ = curve_fit(func_polynomial_ratio, x, y, p0=[1, 1, 1])
            yfit = func_polynomial_ratio(x, *parametros)
        elif funcion_seleccionada == "Función logarítmica":
            parametros, _ = curve_fit(func_logarithmic, x, y, p0=[1, 1])
            yfit = func_logarithmic(x, *parametros)
        elif funcion_seleccionada == "Función lineal con producto de coeficientes":
            parametros, _ = curve_fit(func_linear_product, x, y)
            yfit = func_linear_product(x, *parametros)
        elif funcion_seleccionada == "Función cuadrática inversa":
            parametros, _ = curve_fit(func_inverse_quadratic, x, y)
            yfit = func_inverse_quadratic(x, *parametros)
        elif funcion_seleccionada == "Función polinomial inversa":
            parametros, _ = curve_fit(func_inverse_polynomial, x, y)
            yfit = func_inverse_polynomial(x, *parametros)

        # Calcular el coeficiente de determinación R²
        R2 = r2_score(y, yfit)

        # Graficar los datos y el modelo ajustado
        plt.figure(figsize=(10, 6))
        plt.scatter(x, y, label="Datos originales", color='blue')
        plt.plot(x, yfit, label="Modelo ajustado", color='orange')
        plt.xlabel(x_variable)
        plt.ylabel(y_variable)
        plt.title('Ajuste de Regresión No Lineal')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        # Mostrar la gráfica en Streamlit
        st.pyplot(plt)

        # Mostrar resultados
        ecuacion = mostrar_ecuacion(funcion_seleccionada, parametros)
        st.markdown(f"**Modelo seleccionado**: {funcion_seleccionada}")
        st.write("Ecuación del modelo ajustado:", ecuacion)
        st.write("Coeficientes:", parametros)
        st.write("Coeficiente de determinación (R²): {:.4f}".format(R2))

    except Exception as e:
        st.error(f"Ocurrió un error: {e}")
```
