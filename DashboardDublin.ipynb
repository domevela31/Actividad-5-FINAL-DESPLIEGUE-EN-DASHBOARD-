```python
%%writefile app.py 
# Creamos el archivo de la APP en el interprete principal (Python)

############################# IMPLEMENTACI√ìN DE DASHBOARD ################################

# Importamos librer√≠as
from funpymodeling.exploratory import freq_tbl 
import streamlit as st
import plotly.express as px
import pandas as pd
import seaborn as sns  # Importa seaborn para los gr√°ficos
import matplotlib.pyplot as plt  # Importa matplotlib para mostrar el gr√°fico de calor
from sklearn.linear_model import LinearRegression
import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
#################################################################
@st.cache_resource
def load_data():
    df1 = pd.read_csv("Dublin.csv")
    return df1

df1 = load_data()
##########################################################################################
# Agregar imagen en la barra lateral
image = "https://seeklogo.com/images/D/dublin-logo-30EF152DC2-seeklogo.com.png"
st.sidebar.image(image, use_column_width=True)


###################### CREACI√ìN DEL DASHBOARD ################################

st.sidebar.title("DASHBOARD_Dublin")
st.sidebar.header("Sidebar")

Frames = st.sidebar.selectbox(label="Frames", options=["An√°lisis Univariado Dublin", "Regresi√≥n lineal Simple", "Regresi√≥n M√∫ltiple", "Regresi√≥n no lineal"])
#########################################################################
if Frames == "An√°lisis Univariado Dublin":
    st.title("An√°lisis Univariado de Variables Categ√≥ricas Dublin")
    
    if st.sidebar.checkbox("Mostrar Dataset"):
        st.write(df1)

    Variables = st.sidebar.selectbox("Variable categ√≥rica", df1.select_dtypes(include=['object']).columns)
    Vars_Num = st.selectbox("M√©trica a graficar", df1.select_dtypes(include=['float', 'int']).columns)
    tipo_grafico = st.sidebar.selectbox("Tipo de gr√°fico", ["Barras", "L√≠neas", "Dispersi√≥n", "Circular"])
    vars_multiselect = st.sidebar.multiselect("Selecciona m√∫ltiples m√©tricas a graficar", df1.select_dtypes(include=['float', 'int']).columns)

    if st.sidebar.button("Actualizar Gr√°fico"):
        if tipo_grafico == "Barras":
            figure = px.bar(data_frame=df1, x=Variables, y=Vars_Num)
        elif tipo_grafico == "L√≠neas":
            figure = px.line(data_frame=df1, x=Variables, y=Vars_Num)
        elif tipo_grafico == "Dispersi√≥n":
            figure = px.scatter(data_frame=df1, x=Variables, y=Vars_Num)
        elif tipo_grafico == "Circular":
            figure = px.pie(data_frame=df1, names=Variables)

        st.plotly_chart(figure)

        for var in vars_multiselect:
            if tipo_grafico == "Barras":
                fig_multi = px.bar(data_frame=df1, x=Variables, y=var)
            elif tipo_grafico == "L√≠neas":
                fig_multi = px.line(data_frame=df1, x=Variables, y=var)
            elif tipo_grafico == "Dispersi√≥n":
                fig_multi = px.scatter(data_frame=df1, x=Variables, y=var)
            elif tipo_grafico == "Circular":
                fig_multi = px.pie(data_frame=df1, names=Variables)

            st.plotly_chart(fig_multi)
#########################################################################
if Frames == "Regresi√≥n lineal Simple":
    st.title("Regresi√≥n lineal Simple Dublin")

    # Selecci√≥n de variables independientes y dependientes
    vars_indep = st.multiselect("Selecciona variables independientes", df1.select_dtypes(include=['float', 'int']).columns)
    var_dep = st.selectbox("Selecciona variable dependiente", df1.select_dtypes(include=['float', 'int']).columns)

    # Checkbox para mostrar el conjunto de datos
    if st.checkbox("Mostrar Dataset"):
        st.write(df1)

    # Bot√≥n para realizar la regresi√≥n
    if st.button("Ejecutar Regresi√≥n"):
        # Verificamos que se hayan seleccionado variables
        if len(vars_indep) == 0:
            st.warning("Por favor, selecciona al menos una variable independiente.")
        else:
            # Definimos las variables para la regresi√≥n
            Vars_Indep = df1[vars_indep]
            Var_Dep = df1[var_dep]

            # Limpiamos el DataFrame para la correlaci√≥n
            df_clean = df1[vars_indep + [var_dep]].dropna()  # Solo seleccionamos las columnas necesarias y eliminamos nulos

            # Importamos la clase de regresi√≥n lineal
            from sklearn.linear_model import LinearRegression
            model = LinearRegression()

            # Ajustamos el modelo
            model.fit(X=Vars_Indep, y=Var_Dep)

            # Mostramos los coeficientes
            coef = model.coef_
            intercept = model.intercept_
            st.write("Coeficientes:", coef)
            st.write("Intercepto:", intercept)

            # Evaluamos el modelo
            r_squared = model.score(Vars_Indep, Var_Dep)
            st.write("R¬≤:", r_squared)

            # Predicciones
            y_pred = model.predict(X=Vars_Indep)
            df1['Predicciones'] = y_pred

            # Gr√°fico de dispersi√≥n
            fig = px.scatter(df1, x=vars_indep[0], y=var_dep, title='Gr√°fico de Dispersi√≥n')
            st.plotly_chart(fig)

            # Gr√°fico de calor de correlaci√≥n
            Corr_Factors1 = df_clean.corr()
            Heat_Map = sns.heatmap(Corr_Factors1, cmap='coolwarm', annot=True)
            st.pyplot(fig=Heat_Map.figure)
###########################################################################################
if Frames == "Regresi√≥n M√∫ltiple":
    st.title("Regresi√≥n M√∫ltiple Dublin")

    # Selecci√≥n de variables independientes y dependientes
    vars_indep = st.multiselect("Selecciona variables independientes", df1.select_dtypes(include=['float', 'int']).columns)
    var_dep = st.selectbox("Selecciona variable dependiente", df1.select_dtypes(include=['float', 'int']).columns)

    # Checkbox para mostrar el conjunto de datos
    if st.checkbox("Mostrar Dataset"):
        st.write(df1)

    # Bot√≥n para realizar la regresi√≥n
    if st.button("Ejecutar Regresi√≥n"):
        # Verificamos que se hayan seleccionado variables
        if len(vars_indep) == 0:
            st.warning("Por favor, selecciona al menos una variable independiente.")
        else:
            # Definimos las variables para la regresi√≥n
            Vars_Indep = df1[vars_indep]
            Var_Dep = df1[var_dep]

            # Limpiamos el DataFrame para la correlaci√≥n
            df_clean = df1[vars_indep + [var_dep]].dropna()  # Solo seleccionamos las columnas necesarias y eliminamos nulos

            # Ajustamos el modelo de regresi√≥n m√∫ltiple
            model = LinearRegression()
            model.fit(X=Vars_Indep, y=Var_Dep)

            # Mostramos los coeficientes
            coef = model.coef_
            intercept = model.intercept_
            st.write("Coeficientes:", coef)
            st.write("Intercepto:", intercept)

            # Evaluamos el modelo
            r_squared = model.score(Vars_Indep, Var_Dep)
            st.write("R¬≤:", r_squared)

            # Predicciones
            y_pred = model.predict(X=Vars_Indep)
            df1['Predicciones'] = y_pred

            # Gr√°ficos de dispersi√≥n para cada variable independiente con color
            for var in vars_indep:
                fig = px.scatter(df1, x=var, y=var_dep,
                                 title=f'Gr√°fico de Dispersi√≥n: {var} vs {var_dep}',
                                 color=var_dep,  # Colorear seg√∫n la variable dependiente
                                 color_continuous_scale=px.colors.sequential.Viridis,  # Paleta de colores
                                 size_max=10)  # Ajustar el tama√±o m√°ximo de los puntos
                st.plotly_chart(fig)

                # Agregar la l√≠nea de regresi√≥n al gr√°fico de dispersi√≥n
                fig.add_traces(px.scatter(df1, x=var, y=y_pred, trendline="ols").data)

            # Gr√°fico de calor de correlaci√≥n
            Corr_Factors1 = df_clean.corr()
            plt.figure(figsize=(10, 6))  # Ajusta el tama√±o de la figura
            Heat_Map = sns.heatmap(Corr_Factors1, cmap='coolwarm', annot=True)
            st.pyplot(fig=Heat_Map.figure)
##################################################################################
# Definir funciones no lineales
def func_quadratic(x, a, b, c):
    return a * x**2 + b * x + c

def func_exponential(x, a, b, c):
    return a * np.exp(b * x) + c

def func_inverse(x, a):
    return 1 / a * x

def func_sine(x, a, b):
    return a * np.sin(x) + b

def func_tangent(x, a, b):
    return a * np.tan(x) + b

def func_absolute(x, a, b, c):
    return a * np.abs(x) + b * x + c

def func_polynomial_ratio(x, a, b, c):
    return (a * x**2 + b) / (c * x)

def func_logarithmic(x, a, b):
    return a * np.log(x) + b

def func_linear_product(x, a, b, c):
    return a * x + b * x + c * x

def func_inverse_quadratic(x, a):
    return 1 / a * x**2

def func_inverse_polynomial(x, a, b, c):
    return a / b * x**2 + c * x

def mostrar_ecuacion(funcion, parametros):
    if funcion == "Funci√≥n cuadr√°tica":
        return f"y = {parametros[0]:.4f}x¬≤ + {parametros[1]:.4f}x + {parametros[2]:.4f}"
    elif funcion == "Funci√≥n exponencial":
        return f"y = {parametros[0]:.4f}e^({parametros[1]:.4f}x) + {parametros[2]:.4f}"
    elif funcion == "Funci√≥n inversa":
        return f"y = 1 / ({parametros[0]:.4f}) * x"
    elif funcion == "Funci√≥n senoidal":
        return f"y = {parametros[0]:.4f}sin(x) + {parametros[1]:.4f}"
    elif funcion == "Funci√≥n tangencial":
        return f"y = {parametros[0]:.4f}tan(x) + {parametros[1]:.4f}"
    elif funcion == "Funci√≥n valor absoluto":
        return f"y = {parametros[0]:.4f}|x| + {parametros[1]:.4f}x + {parametros[2]:.4f}"
    elif funcion == "Funci√≥n cociente entre polinomios":
        return f"y = ({parametros[0]:.4f}x¬≤ + {parametros[1]:.4f}) / ({parametros[2]:.4f}x)"
    elif funcion == "Funci√≥n logar√≠tmica":
        return f"y = {parametros[0]:.4f}ln(x) + {parametros[1]:.4f}"
    elif funcion == "Funci√≥n lineal con producto de coeficientes":
        return f"y = {parametros[0]:.4f}x + {parametros[1]:.4f}x + {parametros[2]:.4f}x"
    elif funcion == "Funci√≥n cuadr√°tica inversa":
        return f"y = 1 / ({parametros[0]:.4f}) * x¬≤"
    elif funcion == "Funci√≥n polinomial inversa":
        return f"y = ({parametros[0]:.4f} / {parametros[1]:.4f}) * x¬≤ + {parametros[2]:.4f}x"

if Frames == "Regresi√≥n no lineal":
    st.title("Regresi√≥n no lineal Dublin")

    # Selecci√≥n de variables independientes y dependientes
    x_variable = st.selectbox('Selecciona la variable independiente (x)', df1.select_dtypes(include=['float', 'int']).columns)
    y_variable = st.selectbox('Selecciona la variable dependiente (y)', df1.select_dtypes(include=['float', 'int']).columns)

    # Aseg√∫rate de que los datos seleccionados no tengan valores NaN
    Vars_Indep = df1[[x_variable]].dropna()
    Var_Dep = df1[y_variable].dropna()

    # Redefinir las variables para el ajuste
    x = Vars_Indep[x_variable].values
    y = Var_Dep.values

    # Crear un selector para elegir la funci√≥n de ajuste
    funcion_seleccionada = st.selectbox('Selecciona la funci√≥n de ajuste', 
        ["Funci√≥n cuadr√°tica", "Funci√≥n exponencial", "Funci√≥n inversa", 
         "Funci√≥n senoidal", "Funci√≥n tangencial", "Funci√≥n valor absoluto", 
         "Funci√≥n cociente entre polinomios", "Funci√≥n logar√≠tmica", 
         "Funci√≥n lineal con producto de coeficientes", 
         "Funci√≥n cuadr√°tica inversa", "Funci√≥n polinomial inversa"])

    # Ajustar el modelo seg√∫n la funci√≥n seleccionada
    try:
        if funcion_seleccionada == "Funci√≥n cuadr√°tica":
            parametros, _ = curve_fit(func_quadratic, x, y)
            yfit = func_quadratic(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n exponencial":
            parametros, _ = curve_fit(func_exponential, x, y, p0=[1, 0.1, 1])
            yfit = func_exponential(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n inversa":
            parametros, _ = curve_fit(func_inverse, x, y)
            yfit = func_inverse(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n senoidal":
            parametros, _ = curve_fit(func_sine, x, y)
            yfit = func_sine(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n tangencial":
            parametros, _ = curve_fit(func_tangent, x, y)
            yfit = func_tangent(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n valor absoluto":
            parametros, _ = curve_fit(func_absolute, x, y)
            yfit = func_absolute(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n cociente entre polinomios":
            parametros, _ = curve_fit(func_polynomial_ratio, x, y, p0=[1, 1, 1])
            yfit = func_polynomial_ratio(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n logar√≠tmica":
            parametros, _ = curve_fit(func_logarithmic, x, y, p0=[1, 1])
            yfit = func_logarithmic(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n lineal con producto de coeficientes":
            parametros, _ = curve_fit(func_linear_product, x, y)
            yfit = func_linear_product(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n cuadr√°tica inversa":
            parametros, _ = curve_fit(func_inverse_quadratic, x, y)
            yfit = func_inverse_quadratic(x, *parametros)
        elif funcion_seleccionada == "Funci√≥n polinomial inversa":
            parametros, _ = curve_fit(func_inverse_polynomial, x, y)
            yfit = func_inverse_polynomial(x, *parametros)

        # Calcular el coeficiente de determinaci√≥n R¬≤
        R2 = r2_score(y, yfit)

        # Graficar los datos y el modelo ajustado
        plt.figure(figsize=(10, 6))
        plt.scatter(x, y, label="Datos originales", color='blue')
        plt.plot(x, yfit, label="Modelo ajustado", color='orange')
        plt.xlabel(x_variable)
        plt.ylabel(y_variable)
        plt.title('Ajuste de Regresi√≥n No Lineal')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        # Mostrar la gr√°fica en Streamlit
        st.pyplot(plt)

        # Mostrar resultados
        ecuacion = mostrar_ecuacion(funcion_seleccionada, parametros)
        st.markdown(f"**Modelo seleccionado**: {funcion_seleccionada}")
        st.write("Ecuaci√≥n del modelo ajustado:", ecuacion)
        st.write("Coeficientes:", parametros)
        st.write("Coeficiente de determinaci√≥n (R¬≤): {:.4f}".format(R2))

    except Exception as e:
        st.error(f"Ocurri√≥ un error: {e}")
```
